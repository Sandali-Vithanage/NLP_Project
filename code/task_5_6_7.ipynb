```python
from google.cloud import vision
import io
import pandas as pd

def detect_image_labels(image_path):
    # Initialize the Vision client
    # client = vision.ImageAnnotatorClient()
    client = vision.ImageAnnotatorClient(
        client_options={"quota_project_id": "nlp-group-project-440615"}
    )


    with io.open(image_path, 'rb') as image_file:
        content = image_file.read()
    image = vision.Image(content=content)

    # Detect labels
    response = client.label_detection(image = image)
    labels = [(label.description, label.score) for label in response.label_annotations]

    return labels



```

```python
credentials_path = '../credentials/big-star-collectibles-424408-5fb21f40e30a.json'
if not credentials_path:
    raise ValueError("Google Application Credentials not found.")

# Load the CSV file
file_path = '../dataset/processed/multimodal_dataset.csv'
data = pd.read_csv(file_path, encoding='utf-8')

image_column = 'image_path'

data['image_labels'] = None

for index, row in data.iterrows():
    image_path = row[image_column]
    print(f'image_path:{image_path}')
    labels = detect_image_labels(image_path)
    data.at[index, 'image_labels'] = labels

# Save results to a new csv
output_path = '../dataset/processed/multimodal_dataset_image_labels.csv'
data.to_csv(output_path)
print(f"Labeled images saved to {output_path}")
```

```python
import spacy
text_column = data['text']

# Load spaCy's English language model
nlp = spacy.load("en_core_web_sm")

entities = []
for text in text_column:
    doc = nlp(text)  # Process each text entry
    entities.append([(ent.text, ent.label_) for ent in doc.ents])  # Collect entities

# Add the entities as a new column in the DataFrame
data['text_entities'] = entities

# Display the DataFrame with extracted entities
print(data['text_entities'])
```

Task 7: Fuzzy string matching and overlap percentage Objective: Find the similarity between the image labels and extracted entities using FuzzyWuzzy library. Later calculate a percentage overlap score using the formula above flow chart.

```python
unique_image_labels = set()
for labels in data['image_labels']:
    for label, _ in labels:
        unique_image_labels.add(label)

unique_image_labels = list(unique_image_labels) 
unique_image_labels
```

```python
unique_text_entities = set()
for entities in data['text_entities']:
    for entity, _ in entities:
        unique_text_entities.add(entity)

unique_text_entities = list(unique_text_entities) 
unique_text_entities
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

threshold = 60
matches = []
for text_entity in unique_text_entities:
    print(f"text_entity:{text_entity}")
    best_match, best_score = process.extractOne(text_entity, unique_image_labels, scorer=fuzz.ratio)
    if best_score >= threshold:
        matches.append((text_entity,best_match, best_score))

# Calculate overlap percentage
overlap_percentage = (len(matches) / len(unique_text_entities)) * 100

# Output results
print("Matches:", matches)
print("Overlap Percentage:", overlap_percentage)
```
